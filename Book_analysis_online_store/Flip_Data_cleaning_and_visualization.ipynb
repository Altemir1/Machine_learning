```python
import pandas as pd

df = pd.read_csv("flip_book_messy.csv")

df
```

#Checking common information about data frame


```python
df.info()
```

#Deleting irrelevant features

```python
# Deleting columns that have 2000 null values or more
threshold = 1000
df = df.dropna(axis=1, thresh=len(df) - threshold)

# Deleting unnecessary features
columns_to_drop = ["Unnamed: 0", "ISBN", "Код товара"]
df.drop(columns=columns_to_drop, inplace=True)

df.info()
```

#Data preprocessing

Renaming columns

```python

column_name_mapping = {'Издательство': 'publishing_house',
                       'Переплет': 'binding',
                       "Количество страниц": "number_of_pages",
                       "Бумага": "paper_type",
                       "Язык": "language",
                       "Дата выхода": "release_date",
                       "Высота издания": "height",
                       "Ширина издания" : "width",
                       "Иллюстрированная": "illustrated",
                       "Формат книги": "format",
                       "Толщина издания": "thickness"}


df = df.rename(columns=column_name_mapping)

df.head()
```

```python
df.isnull().sum()
```

Removing string values from hieght, width and thickness features

```python
columns_to_process = ['height', 'width', 'thickness']

for col in columns_to_process:
    df[col] = df[col].str.replace('мм', '').astype(float)

df.head()
```

Filling numeric features with its mean

```python
columns_to_fill = ["rating", "review_count", "number_of_pages", "height", "width", "thickness"]
for col in columns_to_fill:
    mean_value = df[col].mean()
    df[col] = df[col].fillna(mean_value)

df.head()
```

Filling categorical features

```python
# Filling NaN values with zeros to replace to the values furtherly
df["binding"].replace(0, "твердый переплет", inplace=True)
```

```python
substrings_to_replace = ['твердый переплет', 'твердая обложка']

for substring in substrings_to_replace:
    mask = df['binding'].str.contains(substring)
    df.loc[mask, 'binding'] = 'твердый переплет'

df["binding"].value_counts()
```

```python
df["release_date"].fillna(0, inplace=True)
mean = df["release_date"].mean()
df["release_date"].replace(0, mean, inplace=True)
```

```python
columns_to_fill = ["language", "publishing_house", "paper_type"]

for column in columns_to_fill:
  df[column].fillna(0, inplace=True)
  mode = df[column].mode().iloc[0]
  df[column].replace(0, mode, inplace=True)

df.isnull().sum()
```

```python
mask = df['paper_type'].str.contains("пухлая")
df.loc[mask, 'paper_type'] = 'пухлая'

mask = df['paper_type'].str.contains("мелованная")
df.loc[mask, 'paper_type'] = 'матовая'

types_offset = ["офсетная", "белая", "офсет"]
for type_ in types_offset:
  mask = df['paper_type'].str.contains(type_)
  df.loc[mask, 'paper_type'] = 'оффсетная'

df.loc[df['binding'] == 'мягкий обложка', 'paper_type'] = 'газетная'

df["paper_type"].value_counts()
```

```python
df.to_csv("flip_book_cleaned.csv")
```
