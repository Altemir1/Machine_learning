#Downloading csv file and turn it to dataframe

```python
import pandas as pd
import numpy as np


df = pd.read_csv("almaty_apartments_main.csv")

df
```

#Checking for duplicate rows and deleting them

```python

#
# Check for duplicate rows
duplicate_rows = df[df.duplicated()]

# Print duplicate rows
print("Duplicate Rows except first occurrence:")
print(duplicate_rows)

# Remove duplicate rows
df = df.drop_duplicates()

# Print the DataFrame after removing duplicates
print("DataFrame after removing duplicates:")
print(df)
```

#Converting values in column to right format

```python
# Take number of the floor
df['floor'] = df['floor'].str.split(' ').str[0]

# Delete the area of kitchen from the common area
df["area"] = df["area"].str.split(" ").str[0]

# Deleting "〒" element from price column
df['price'] = df['price'].str.replace(r'\D', '')

# Keeping only number in column number of rooms
df["number_of_rooms"] = df["number_of_rooms"].str.replace("-комнатная квартира", "")

df
```

#Visualizing data

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualizing realtionship between
columns_to_scatter = ["quality", "bathroom", "internet_type"]

# Year to columns to scatter
for column in columns_to_scatter:

  sns.scatterplot(x='year_of_construction', y=column, data=df, hue=column)

  plt.xlabel('Year')
  plt.ylabel(f'{column}')
  plt.title(f'Scatter Plot between Year and {column} columns')

  plt.show()
```

#Identifying and filling null values


```python
null_values = df.isnull().sum()

# Print the number of null values in each column
print("Number of null values in each column:")
print(null_values)
```

```python
# Delete row where column price is null
df.dropna(subset=['price'], inplace=True)

# Fill district, structure_type, floor
columns_to_fill = ['district', 'structure_type', 'floor']

# Fill null values in specific columns with their modes
for column in columns_to_fill:

    mode_value = df[column].mode()[0]
    df[column].fillna(mode_value, inplace=True)

# Fill internet type with diferent values

random_values_internet = np.random.choice(["через TV кабель", "ADSL", "оптика", "проводной"], size=len(df), replace=True)
df['internet_type'].fillna(pd.Series(random_values_internet, index=df.index), inplace=True)

# Fill null value in residential_complex with no rc
df["residential_complex"].fillna("not in RC", inplace=True)

# Fill bathroom type
random_values_bathroom = np.random.choice(["совмещенный", "раздельный", "2 с/у и более"], size=len(df), replace=True)
df['bathroom'].fillna(pd.Series(random_values_bathroom, index=df.index), inplace=True)

# Fill quality
# Filling new building with quality черновая отделка
df.loc[df['year_of_construction'] == 2024, 'quality'] = "черновая отделка"

# Fill with values средний and  хорошое
random_values_quality = np.random.choice(["хорошее", "среднее"], size=len(df), replace=True)
df['quality'].fillna(pd.Series(random_values_quality, index=df.index), inplace=True)


```
